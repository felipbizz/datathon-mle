```python
import os

if os.path.basename(os.getcwd()) == "notebooks":
    os.chdir("..")
```

```python
import pandas as pd
from mle_datathon.utils import (
    get_abs_path, 
    load_config, 
    set_log
)

local_path = os.getcwd()
config = load_config(local_path)
paths = config["paths"]

df = pd.read_parquet(get_abs_path(local_path, paths["dataset_features"]))
df
```

```python
features = [
        col
        for col in df.columns
        if col != "target" and pd.api.types.is_numeric_dtype(df[col])
    ]

df = df[features]

caso_de_uma_pessoa_aleatoria = df.sample(1).values[0]
caso_de_uma_pessoa_aleatoria
```

```python
import pickle

with open(paths["modelo_treinado"], "rb") as f:
    dados_carregados = pickle.load(f)

modelo = dados_carregados["model"]
imputer = dados_carregados["imputer"]
scaler = dados_carregados["scaler"]
features = dados_carregados["features"]
```

```python
caso_imputer = imputer.transform([caso_de_uma_pessoa_aleatoria])[0]
caso_scaler = scaler.transform([caso_imputer])[0]
modelo.predict_proba([caso_scaler])
```
